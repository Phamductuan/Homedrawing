```python
import turtle
import math
# tô màu nền
t= turtle. Turtle()
turtle.bgcolor("#A3E4D7")
t.speed(9)
# Xay dung ham ve hinh chu nhat
def drawrectangle(m, width, height, color):
    t.fillcolor(color)
    t.begin_fill()
    for i in range(2):
        t.fd(width)
        t.lt(90)
        t.fd(height)
        t.lt(90)
    t.end_fill() 
# xay dung ham ve hinh tam giac
def drawtriangle(m,length,color):
    t.fillcolor(color)
    t.begin_fill()
    t.fd(length)
    t.lt(135)
    t.fd(length/math.sqrt(2))
    t.lt(90)
    t.fd(length/math.sqrt(2))
    t.lt(135)
    t.end_fill()
# Xay dung ham ve hinh binh hanh
def drawparallelogram(m,width,height,color):
    t.fillcolor(color)
    t.begin_fill()
    t.lt(30)
    t.fd(width)
    t.lt(60)
    t.fd(height)
    t.lt(120)
    t.fd(width)
    t.lt(60)
    t.fd(height)
    t.lt(90)
    t.end_fill()
# mat truoc ngoi nha
t.penup();t.goto(-150,-120);t.pendown()
drawrectangle(t,100,110,"blue")
# cua ra vao
t.fd(30);drawrectangle(t,40,60,"#76D7C4")
# ve mai nha tam giac
t.penup(); t.goto(-150,-10);t.pendown()
drawtriangle(t,100,"pink")
# ve canh ngoi nha
t.penup();t.goto(-50,-120);t.pendown()
drawparallelogram(t,60,110,"yellow")
# ve cua so
t.penup(); t.goto(-30,-60);t.pendown()
drawparallelogram(t,20,30,"brown")
# ve mai nha hinh binh hanh
t.fillcolor("#F8C471");t.begin_fill()
t.penup();t.goto(-50,-10);t.pendown()
t.lt(30);t.fd(60);t.lt(105);t.fd(100/math.sqrt(2));t.lt(75);t.fd(60);t.lt(105);t.fd(100/math.sqrt(2))
t.end_fill()
# ve tree
t.penup();t.goto(150,-120);t.pendown();t.lt(45)
drawrectangle(t,20,40,"#F0B27A")
# ve cay tam giac
t.penup();t.goto(120,-80);t.pendown()
drawtriangle(t,80,"#76D7C4")
t.penup();t.goto(130,-40);t.pendown()
drawtriangle(t,60,"#76D7C4")
t.penup();t.goto(140,-10);t.pendown()
drawtriangle(t,40,"#76D7C4")
# xay dung ham ve sun rays
def drawsunray(m,length, radius):
    for i in range(4):
        t.penup()
        t.fd(radius)
        t.pendown()
        t.fd(length)
        t.penup()
        t.bk(length + radius)
        t.pendown()
        t.lt(90)
t.fillcolor("yellow");t.begin_fill()
t.penup();t.goto(100,90);t.pendown()
# ve ong mat troi
t.circle(24)
t.end_fill()
t.penup();t.goto(100,114);t.pendown()
drawsunray(t,25,24)
t.lt(45)
drawsunray(t,15,24)
t.penup();t.goto(-150,-150);t.pendown()
turtle.shape("turtle");turtle.penup();turtle.goto(-150,-150);turtle.pendown();turtle.lt(90)
turtle.done()
```
